```python
#task 3 12
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Assuming you have loaded the Pima Indians Diabetes dataset into a DataFrame named 'diabetes_data'
# Replace 'your_dataset.csv' with the actual file path or URL if necessary.

# Load the dataset
diabetes_data= pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# Explore correlation and scatter plots
correlation_matrix = diabetes_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Histograms
diabetes_data.hist(figsize=(10, 10))
plt.suptitle("Histograms of Features")
plt.show()

# Three-dimensional plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(diabetes_data['feature1'], diabetes_data['feature2'], diabetes_data['target_variable'], c='r', marker='o')
ax.set_xlabel('feature1')
ax.set_ylabel('feature2')
ax.set_zlabel('Target Variable')
plt.title("Three-dimensional Scatter Plot")
plt.show()
```

```python
#task 1 7
import pandas as pd
import numpy as np

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# Display the first few rows of the dataset
print(df.head())

# Frequency
frequency = df['Glucose'].value_counts()
print("Frequency:\n", frequency)

# Mean
mean = df['Glucose'].mean()
print("\nMean:", mean)

# Median
median = df['Glucose'].median()
print("Median:", median)

# Mode
mode = df['Glucose'].mode()
print("Mode:\n", mode)

# Variance
variance = df['Glucose'].var()
print("\nVariance:", variance)

# Standard Deviation
std_deviation = df['Glucose'].std()
print("Standard Deviation:", std_deviation)

# Skewness
skewness = df['Insulin'].skew()
print("\nSkewness:", skewness)

# Kurtosis
kurtosis = df['Insulin'].kurt()
print("Kurtosis:", kurtosis)
```

```python
#task 1 9
import pandas as pd
import statsmodels.api as sm

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# Replace 'TargetColumn' with the column you want to predict and 'PredictorColumn1', 'PredictorColumn2', etc., with your predictor variables
X = df[['Pregnancies', 'Glucose', 'BloodPressure']]
y = df['Glucose']

# Add a constant term to the predictors
X = sm.add_constant(X)

# Fit the multiple regression model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```
