```python

import pandas as pd

# Assuming df is your DataFrame
# You can create a DataFrame for demonstration purposes
data = {'Column1': [1, 2, 3, 4, 5],
        'Column2': ['A', 'B', 'C', 'D', 'E'],
        'Column3': [10.5, 20.3, 15.8, 7.2, 12.1]}

df = pd.DataFrame(data)

# Display the first n rows of the DataFrame
print("First 3 rows:")
print(df.head(3))

# Display the last n rows of the DataFrame
print("\nLast 3 rows:")
print(df.tail(3))

# Display concise summary of the DataFrame
print("\nDataFrame Info:")
print(df.info())

# Generate descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Returns the dimensions of the DataFrame
print("\nDataFrame Shape:")
print(df.shape)

# Returns the column labels
print("\nColumn Labels:")
print(df.columns)

# Returns the data types of each column
print("\nData Types of Columns:")
print(df.dtypes)
```

```python

import pandas as pd

# Assuming df is your DataFrame
# You can create a DataFrame for demonstration purposes
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

# Select a single column
single_column = df['Age']
print("Single Column:")
print(single_column)

# Select multiple columns
multiple_columns = df[['Name', 'City']]
print("\nMultiple Columns:")
print(multiple_columns)

# Access rows and columns by labels using loc
# In this example, we select the first two rows and the 'Name' column
loc_selection = df.loc[:1, 'Name']
print("\nSelection using loc:")
print(loc_selection)

# Access rows and columns by integer positions using iloc
# In this example, we select the first two rows and the first column
iloc_selection = df.iloc[:2, 0]
print("\nSelection using iloc:")
print(iloc_selection)
```

```python

import pandas as pd
import numpy as np

# Assuming df is your DataFrame
# You can create a DataFrame for demonstration purposes
data = {'Name': ['Alice', 'Bob', 'Charlie', np.nan],
        'Age': [25, np.nan, 22, 30],
        'City': ['New York', 'San Francisco', np.nan, 'Los Angeles']}

df = pd.DataFrame(data)

# Check for missing values
print("Check for Missing Values:")
print(df.isnull())

# Check for non-missing values
print("\nCheck for Non-Missing Values:")
print(df.notnull())

# Drop rows with missing values
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with missing values:")
print(df_dropped)

# Fill missing values with a specific value, e.g., fill NaNs in 'Age' with the mean value
df_filled = df.fillna({'Age': df['Age'].mean()})
print("\nDataFrame after filling missing values:")
print(df_filled)
```

```python

import pandas as pd

# Assuming df is your DataFrame
# You can create a DataFrame for demonstration purposes
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob'],
        'Age': [25, 30, 22, 28, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'New York', 'San Francisco'],
        'Salary': [50000, 60000, 48000, 55000, 65000]}

df = pd.DataFrame(data)

# Group DataFrame by 'Name'
grouped_by_name = df.groupby('Name').mean()
print("Grouped by Name:")
print(grouped_by_name)

# Sort DataFrame by 'Age' in descending order
sorted_df = df.sort_values(by='Age', ascending=False)
print("\nSorted by Age:")
print(sorted_df)

# Create a pivot table with 'Name' as index, 'City' as columns, and 'Salary' as values
pivot_table = df.pivot_table(values='Salary', index='Name', columns='City', aggfunc='sum', fill_value=0)
print("\nPivot Table:")
print(pivot_table)

# Unpivot the DataFrame using melt
melted_df = pd.melt(df, id_vars=['Name'], value_vars=['Age', 'Salary'], var_name='Attribute', value_name='Value')
print("\nUnpivoted DataFrame:")
print(melted_df)
```

```python

import pandas as pd

# Assuming df1 and df2 are your DataFrames
# You can create DataFrames for demonstration purposes
data1 = {'ID': [1, 2, 3],
         'Name': ['Alice', 'Bob', 'Charlie']}
df1 = pd.DataFrame(data1)

data2 = {'ID': [3, 4, 5],
         'City': ['New York', 'San Francisco', 'Los Angeles']}
df2 = pd.DataFrame(data2)

# Concatenate DataFrames along rows (axis=0)
concatenated_df = pd.concat([df1, df2], axis=0, ignore_index=True)
print("Concatenated DataFrame:")
print(concatenated_df)

# Merge DataFrames based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print("\nMerged DataFrame:")
print(merged_df)
```
